. How does the applicant's income type affect the loan amount and approval rate?

 Insight: Understanding how different income types influence loan amounts and approvals can guide personalized marketing and risk management.


```python
import pandas as pd

# Assuming the data is stored in a CSV named 'application_data.csv' and loaded into a dataframe named 'dfmer'

# Filter for loans with income type information
df_filtered = dfmer[dfmer['NAME_INCOME_TYPE'].notna()]

# Group data by income type
grouped_data = df_filtered.groupby('NAME_INCOME_TYPE').agg({
  'AMT_CREDIT_in_lakhs': ['mean'],  # Average loan amount
  'TARGET': ['mean']  # Approval rate (converted to mean - 1 represents approval, 0 represents rejection)
})

# Rename columns for better readability
grouped_data.columns = ['Average Loan Amount (in lakhs)', 'Approval Rate']

# Round values for better presentation (optional)
grouped_data['Average Loan Amount (in lakhs)'] = grouped_data['Average Loan Amount (in lakhs)'].round(2)
grouped_data['Approval Rate'] = grouped_data['Approval Rate'].round(2)

# Print results
print(grouped_data)
```

```python
import pandas as pd

# Assuming the data is stored in a CSV named 'application_data.csv' and loaded into a dataframe named 'dfmer'

# Filter for loans with income type information
df_filtered = dfmer[dfmer['NAME_INCOME_TYPE'].notna()]

# Group data by income type
grouped_data = df_filtered.groupby('NAME_INCOME_TYPE').agg({
  'AMT_CREDIT_x': ['mean'],  # Average loan amount (assuming AMT_CREDIT_x is the loan amount)
  'TARGET': ['mean']  # Approval rate (converted to mean - 1 represents approval, 0 represents rejection)
})

# Rename columns for better readability
grouped_data.columns = ['Average Loan Amount (in lakhs)', 'Approval Rate']

# Round values for better presentation (optional)
grouped_data['Average Loan Amount (in lakhs)'] = grouped_data['Average Loan Amount (in lakhs)'].round(2)
grouped_data['Approval Rate'] = grouped_data['Approval Rate'].round(2)

# Print results
print(grouped_data)
```

# Business question

1. What are the approval rates across different loan purposes, and how do they compare with the repayment behavior?
Insight: This analysis can help identify which loan purposes have higher approval and subsequent repayment success rates, aiding in risk assessment and product focus adjustments.


```python
df_mer = pd.merge(df_app, df_prev, on='SK_ID_CURR', suffixes=('_app', '_prev'))

# Calculate approval rates across different loan purposes
approval_rates = df_mer[df_mer['NAME_CONTRACT_STATUS'] == 'Approved']['NAME_CASH_LOAN_PURPOSE'].value_counts(normalize=True) * 100

# Calculate repayment behavior
repayment_behavior = df_mer[df_mer['NAME_CONTRACT_STATUS'] == 'Approved'].groupby('NAME_CASH_LOAN_PURPOSE')['TARGET'].mean() * 100

# Combine the results into a single DataFrame
approval_repayment = pd.DataFrame({'Approval Rate (%)': approval_rates, 'Repayment Rate (%)': 100 - repayment_behavior})

# Plot the results
plt.figure(figsize=(12, 8), facecolor='white')
ax = sns.barplot(x=approval_repayment.index, y='Approval Rate (%)', data=approval_repayment, color='blue', label='Approval Rate')
ax2 = ax.twinx()
ax2 = sns.lineplot(x=approval_repayment.index, y='Repayment Rate (%)', data=approval_repayment, color='red', marker='o', label='Repayment Rate')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Approval and Repayment Rates by Loan Purpose')
ax.set_xlabel('Loan Purpose')
ax.set_ylabel('Approval Rate (%)')
ax2.set_ylabel('Repayment Rate (%)')
plt.legend(loc='upper left')
plt.show()

# Save the combined results to a CSV file
approval_repayment.to_csv('approval_repayment.csv')

print('Approval and repayment rates calculated and plotted.')
```

2. How does the applicant's income type affect the loan amount and approval rate?
 Insight: Understanding how different income types influence loan amounts and approvals can guide personalized marketing and risk management.


```python
# Calculate the average loan amount and approval rate by income type using the correct column names
income_loan_amount = df_mer[df_mer['NAME_CONTRACT_STATUS'] == 'Approved'].groupby('NAME_INCOME_TYPE')['AMT_CREDIT_app'].mean()
income_approval_rate = df_mer[df_mer['NAME_CONTRACT_STATUS'] == 'Approved']['NAME_INCOME_TYPE'].value_counts(normalize=True) * 100

# Combine the results into a single DataFrame
income_analysis = pd.DataFrame({'Average Loan Amount': income_loan_amount, 'Approval Rate (%)': income_approval_rate})

# Plot the results
plt.figure(figsize=(12, 8), facecolor='white')
ax = sns.barplot(x=income_analysis.index, y='Average Loan Amount', data=income_analysis, color='blue', label='Average Loan Amount')
ax2 = ax.twinx()
ax2 = sns.lineplot(x=income_analysis.index, y='Approval Rate (%)', data=income_analysis, color='red', marker='o', label='Approval Rate')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Average Loan Amount and Approval Rate by Income Type')
ax.set_xlabel('Income Type')
ax.set_ylabel('Average Loan Amount')
ax2.set_ylabel('Approval Rate (%)')
plt.legend(loc='upper left')
plt.show()

# Save the combined results to a CSV file
income_analysis.to_csv('income_analysis.csv')

print('Income type analysis calculated and plotted.')
```

```python
def Uni_Analysis_Numerical(dataframe,column):
    sns.set(style='darkgrid')
    plt.figure(figsize=(25,5))
    
    plt.subplot(1,3,1)
    sns.boxplot(data=dataframe,x=column,orient="v").set(title="Box Plot")
    plt.subplot(1,3,2)
    sns.displot(dataframe[column].dropna()).set(title="Displot")
    
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def Uni_Analysis_Numerical(dataframe, column):
    sns.set(style='darkgrid')
    plt.figure(figsize=(25, 5))
    
    plt.subplot(1, 3, 1)
    sns.boxplot(data=dataframe, x=column, orient="v").set(title="Box Plot")
    
    plt.subplot(1, 3, 2)
    sns.histplot(dataframe[column].dropna(), kde=True).set(title="Histogram")
    
    plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def Uni_Analysis_Categorical_BarPlot(dataframe, column):
    """
    Function to perform univariate analysis on a categorical column using a bar plot.
    
    Parameters:
    dataframe (pd.DataFrame): The DataFrame containing the data.
    column (str): The name of the categorical column to analyze.
    """
    sns.set(style='darkgrid')
    plt.figure(figsize=(10, 6))
    
    # Bar Plot
    sns.countplot(data=dataframe, x=column)
    plt.title(f'Bar Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)  # Rotate x labels if needed for better readability
    
    plt.show()

import warnings
warnings.filterwarnings('ignore')

# Example usage:
# Assuming 'df_app' is your DataFrame and 'categorical_data' contains the names of the categorical columns.
for col in categorical_data:
    Uni_Analysis_Categorical_BarPlot(df_app, col)
```

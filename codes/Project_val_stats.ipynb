```python
plt.figure(figsize=[15,6])

region_BARH=df_val['region'].value_counts()
#can also add custom colour this
colors = ['#084594', '#2171b5', '#4292c6', '#6baed6', '#9ecae1', '#c6dbef', '#deebf7', '#f7fbff']
region_BARH.plot.barh(width=0.8,color=colors)

plt.title("region",fontdict={'fontsize':15},pad=10)


for index,value in enumerate(region_BARH):
    plt.text(value,index,f"{value}",va='center',ha='left',color='black',fontsize=10)

plt.show()
```

```python
plt.figure(figsize=[15,6])

rating_BARH=df_val['rating'].value_counts()
#can also add custom colour this
colors = ['#084594', '#2171b5', '#4292c6', '#6baed6', '#9ecae1', '#c6dbef', '#deebf7', '#f7fbff']
rating_BARH.plot.barh(width=0.8,color=colors)

plt.title("rating",fontdict={'fontsize':15},pad=10)


for index,value in enumerate(rating_BARH):
    plt.text(value,index,f"{value}",va='center',ha='left',color='black',fontsize=10)
    
    
plt.show()
```
